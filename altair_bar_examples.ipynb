```python
import altair as alt
alt.renderers.enable('notebook')
alt.data_transformers.enable('default', max_rows=None)
#alt.data_transformers.enable('json')
alt.themes.enable('opaque') # for dark background color
from vega_datasets import data

import pandas as pd
import numpy as np
```

```python
cars = data.cars()
```

```python
cars.head()
```

```python
alt.Chart(data=cars).mark_bar().encode(
    x="Origin",
    y="mean(Horsepower)",
)
```

```python
alt.Chart(data=cars).mark_bar().encode(
    y="Origin",
    x="mean(Horsepower)",
)
```

```python
alt.Chart(data=cars).mark_bar().encode(
    y="Origin",
    x="mean(Horsepower)",
    size="count()",
)
```

```python
alt.Chart(data=cars).mark_bar().encode(
    y="Origin",
    x="mean(Horsepower)",
    size="count()",
    tooltip=["Origin","mean(Horsepower)","count()"],
)
```

```python
alt.Chart(data=cars).mark_bar().encode(
    y="Origin",
    x="mean(Horsepower)",
    size="count()",
    row="year(Year)",
    tooltip=["Origin","mean(Horsepower)","count()"],
)
```

```python
alt.Chart(data=cars).mark_bar().encode(
    row="Origin",
    x="mean(Horsepower)",
    size="count()",
    y="year(Year)",
    tooltip=["Origin","mean(Horsepower)","count()"],
)
```

```python
alt.Chart(data=cars).mark_bar().encode(
    x="Origin",
    y="mean(Horsepower)",
    size="count()",
    column="year(Year)",
    tooltip=["Origin","mean(Horsepower)","count()"],
)
```
